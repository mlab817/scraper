```python
# Create total infrastructure cost
```

```python
cols2016 = ['it_2016_nglocal','it_2016_ngloan','it_2016_odagrant','it_2016_gocc','it_2016_lgu','it_2016_ps','it_2016_others']
df['investment_target_2016'] = df[cols2016].sum(axis=1)
cols2017 = ['it_2017_nglocal','it_2017_ngloan','it_2017_odagrant','it_2017_gocc','it_2017_lgu','it_2017_ps','it_2017_others']
df['investment_target_2017'] = df[cols2017].sum(axis=1)
cols2018 = ['it_2018_nglocal','it_2018_ngloan','it_2018_odagrant','it_2018_gocc','it_2018_lgu','it_2018_ps','it_2018_others']
df['investment_target_2018'] = df[cols2018].sum(axis=1)
cols2019 = ['it_2019_nglocal','it_2019_ngloan','it_2019_odagrant','it_2019_gocc','it_2019_lgu','it_2019_ps','it_2019_others']
df['investment_target_2019'] = df[cols2019].sum(axis=1)
cols2020 = ['it_2020_nglocal','it_2020_ngloan','it_2020_odagrant','it_2020_gocc','it_2020_lgu','it_2020_ps','it_2020_others']
df['investment_target_2020'] = df[cols2020].sum(axis=1)
cols2021 = ['it_2021_nglocal','it_2021_ngloan','it_2021_odagrant','it_2021_gocc','it_2021_lgu','it_2021_ps','it_2021_others']
df['investment_target_2021'] = df[cols2021].sum(axis=1)
cols2022 = ['it_2022_nglocal','it_2022_ngloan','it_2022_odagrant','it_2022_gocc','it_2022_lgu','it_2022_ps','it_2022_others']
df['investment_target_2022'] = df[cols2022].sum(axis=1)
cols2023 = ['it_2023_nglocal','it_2023_ngloan','it_2023_odagrant','it_2023_gocc','it_2023_lgu','it_2023_ps','it_2023_others']
df['investment_target_2023'] = df[cols2023].sum(axis=1)
```
